```python
#droping columns
df=df.drop(columns=['Unnamed: 21','Unnamed: 22','Unnamed: 23','Unnamed: 24','Unnamed: 25'])
print(df.columns)
```

```python
# Handle missing values
# see missing values
print(df.isnull().sum())
```

```python
# filijng missing values in Sales_comminssion_Code with None.
df['sales_commission_code'] = df['sales_commission_code'].fillna('None')

#droping rows in SKU
df = df.dropna(subset=['sku'])

# Filling values in 'M-Y' 'fy' customer ID
df['Customer ID'] = df['Customer ID'].fillna(0)
df['M-Y']=df['M-Y'].fillna('unknown')
df['FY']=df['FY'].fillna('unknown')
```

```python
# checking missing values
df.isnull().sum()
```

```python
# removing the 0 from the price
# Remove rows where price is 0
df = df[df['price'] != 0]

# Check if any zero prices left
print(df['price'].min())
```

```python
# Remove rows where discount is negative
df = df[df['discount_amount'] >= 0]

# Check if any negative discounts left
print(df['discount_amount'].min())
```

```python
# Fixing datatypes
# Convert 'created_at' to datetime format
df['created_at'] = pd.to_datetime(df['created_at'])
print(df.dtypes)
```

**Data Analysis and Business Insights**

```python
# total sales.
total_sales=df['grand_total'].sum()
print(total_sales)













```

```python
# Average sale per order
average_order_value = df['grand_total'].mean()
print(f"Average Order Value: {average_order_value}")
```

```python
# Top 5 customers based on total spend
top_customers = df.groupby('Customer ID')['grand_total'].sum().sort_values(ascending=False).head(5)
print(top_customers)
```

```python
# Top 5 products by revenue
top_products = df.groupby('sku')['grand_total'].sum().sort_values(ascending=False).head(5)
print(top_products)
```

```python
# Sales per year
sales_per_year = df.groupby(df['created_at'].dt.year)['grand_total'].sum()
print(sales_per_year)
```

```python
# Sales per month
sales_per_month = df.groupby(df['created_at'].dt.month)['grand_total'].sum()
print(sales_per_month)
```

```python
# Top customers already calculated
top_customers.plot(kind='bar', figsize=(10,6))
plt.title('Top 5 Customers by Total Spend')
plt.xlabel('Customer ID')
plt.ylabel('Total Spend')
plt.show()
```

```python
# Top products already calculated
top_products.plot(kind='bar', figsize=(10,6))
plt.title('Top 5 Products by Revenue')
plt.xlabel('Product SKU')
plt.ylabel('Revenue')
plt.show()
```

```python
# Sales per year already calculated
sales_per_year.plot(kind='line', marker='o', figsize=(10,6))
plt.title('Sales Per Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# Sales per month already calculated
sales_per_month.plot(kind='bar', figsize=(10,6))
plt.title('Sales Per Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
